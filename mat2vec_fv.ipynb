```python
#to load



loaded_best = load_model('m2v_et')
loaded_best
```

```python
#getting important features and regressing only with 155 of them

feature_importance = loaded_best.feature_importances_

# Create a DataFrame with feature names and their importance scores
feature_importance_df = pd.DataFrame({'Feature': df.drop(['target'], axis = 1).columns, 'Importance': feature_importance})

# Sort the DataFrame by importance scores
top_155_features = feature_importance_df.sort_values(by='Importance', ascending=False).head(155)

# Display the top 155 features
print(top_155_features)

best_features = top_155_features['Feature'].tolist()
```

```python
best_features.append('target')
df[best_features].to_csv('m2v_best155_target.csv')

#doing regression with best_features

#initializing pycaret regression
from pycaret.regression import *
s = setup(df[best_features], target = 'target', session_id = 121, use_gpu = True)

#setting up regression experiment
from pycaret.regression import RegressionExperiment
exp_155 = RegressionExperiment()

#initialize the setup on the experiment 
reg_155 = exp_155.setup(df[best_features], target = 'target', session_id = 121, use_gpu = True)
```

```python
best_155 = reg_155.compare_models()
```

```python
tuned_155 = tune_model(best_155)
```

```python
#residual plot
plot_model(tuned_155, plot = 'residuals')
```

```python
save_model(tuned_155, 'm2v_155')
```
